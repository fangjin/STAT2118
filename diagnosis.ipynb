```r
SAT = c(1263, 1131, 1755, 2070, 1824, 1170, 1245, 1443, 2187, 1503, 1839, 2127, 1098)
GPA = c(2.9, 2.93, 3, 3.45, 3.66, 2.88, 2.15, 2.53, 3.22, 1.99, 2.75, 3.9, 1.6)
reg = lm(GPA ~ SAT)
summary(reg)
```

```r
plot(SAT, GPA)
abline(reg)
```

```r
predict(reg, data.frame(SAT = 1800), interval = "c") 
```

```r
predict(reg, data.frame(SAT = 1800), interval = "p") 
```

```r
cor(SAT, GPA) 
```
